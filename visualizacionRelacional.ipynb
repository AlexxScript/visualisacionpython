```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df.head()
```

### Las gráficas de relación nos permiten explorar relaciones entre dos variables cuales quiera 

**Aquí podemos ver que la propina (tip) sera directamente proporcional al consumo total (total_bill), esto quiere decir que si una variable aumenta la otra también los hace**

```python
sns.set()
sns.relplot(data=df,x="total_bill",y="tip")
```

**Pero que pasa si a partir de esa gráfica quisieramos ver cuantos son del sexo masculino y cuantos del femenino... Pues para divir los datos en diferentes colores tenemos el parámtro "hue" el cual va a dividir los puntos en diferentes colores de acuerdo al número de sexo, es decir, en este momento solo tenemos masculino y femenino, esto quiere decir que los puntos se pintarán en dos colores diferentes**

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="sex")
```

**Ahora la gráfica de arriba nos entrega más información que nos puede ayudar a realizar el análisis, por ejemplo, con la anterior gráfica podemos decir que quienes tienden a gastar más son del sexo masculino.**

**Ahora cambiemos el parámetro de "hue" por time, los puntos siguen siendo del mismo color, sin embargo, cada color nos indica un tiempo de consumo diferente, por lo que podríamos decir que el tiempo en el que se consume más es en la cena (dinner).**

```python
sns.relplot(data=df, x="total_bill", y="tip",hue="time")
```

**Y si quisieramos saber a cuantos le corresponden a cada sexo? Podríamos agregar el parámetro "col" el cual nos va a separar en dos tablas semejantes a la anterior solo que cada tabla corresponde a cada sexo**

```python
sns.relplot(data=df, x="total_bill", y="tip",hue="time",col="sex")
```

**Esto nos entrega aún más información y es que apartir de esas gráficas podemos decir que la mayoría de los hombres tienden a consumir más a la hoa de la cena. Mientras tanto en el caso del sexo femenino podemos ver que es una relación un poco más pareja**
